## Preprocessing

```python
def clear_text(text):
  cleared_text = html.unescape(text)
  cleared_text = cleared_text.split('---------------------------')[1]
  cleared_text = cleared_text.split('\n>')[0]
  cleared_text = cleared_text.replace('\n', '')
  return cleared_text
```

```python
mails['content'] = mails['content'].apply(clear_text)
```
